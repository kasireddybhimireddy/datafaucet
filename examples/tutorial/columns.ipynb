```python
df = dfc.range(1000)
```

```python
df = (df
     .cols.create('öRDER_NUMBER').randint(10)
     .cols.create('REQUEST_ID').randint(10)
     .cols.create('REQUEST_UNUSED_ID').randint(10)
     )
```

```python
(df
    .cols.alter('unidecode', 'lower')
    .cols.rename('id', 'name_id', prefix='search_')
    .cols.find('id').rename(prefix='aaa_')
    .cols.find('unused').drop()
).data.grid(3)
```

```python
df = (df
    .cols.create('Uid').randint(10000000)
    .cols.create('öptïn').randchoice([True, False])
    .cols.create('full$$name').fake('name')
    .cols.create('CREATED_AT').fake('date_between', start_date="-5y", end_date="today")
    .cols.create('лäst_лøgin').fake('past_datetime', start_date="-1d")  
    .cols.create('last-ip0').fake('ipv4_public')      
    .cols.create('gid').randint(10)

).cache()
```

```python
df = df.cols.alter('unidecode', ('translate', '-', '_'), 'alpha', 'lower')
df.data.grid(5)
```
